```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
car = pd.read_csv('C:\\Users\\User\\Downloads\\Automobile_data.csv')
car
```

```python
car.dtypes
```

```python
car.columns
```

```python
#analytical summary of dataset
car.describe(include='all')
```

```python
car.hist(figsize=(20,30))
```

```python
#Relationship between categorical and continous variable
sns.boxplot(x="wheel-base",y="compression-ratio",data=car)
```

```python
sns.pairplot(car)
```

```python
#Drop irrelevant columns 
car= car.drop(['symboling','wheel-base','length','height','curb-weight','engine-size','compression-ratio'],axis=1)
car.head(5)
```

```python
#renaming column names
car= car.rename(columns ={"aspiration":"Alish","width":"widthhhh"})
car.head(5)
```

```python
#Total number of rows and columns
car.shape
```

```python
#rows containing duplicate data
duplicate_rows_car =car[car.duplicated()]
print("Number of duplicate rows:",duplicate_rows_car.shape)
```

```python
#Count the rows before removing the data
car.count()
```

```python
#drop the duplicates
car=car.drop_duplicates()
car.count()
```

```python
# Find the null values
print(car.isnull().sum())
```

```python
#drop the misssing values
car=car.dropna()
car.count()
```

```python
print(car.isnull().sum())
```

```python
#Finding the outliners
sns.boxplot(x=car['height'])
```
